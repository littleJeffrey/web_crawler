```python
import requests
from bs4 import Tag, BeautifulSoup as bs
import pandas as pd
from datetime import datetime
from dateutil import parser
import random

def get_activity(keyword=None):
    if(keyword=="吃了"):
        return "吃了"
    r=requests.get("https://cpe.cse.nsysu.edu.tw/cpe/test_data/problems")
    soup=bs(r.text,"html.parser")
    table=soup.find_all('table',{'class':'table table-bordered table-hover'})
    question_f=[]
    code_f=[]
    sampleA_f=[]
    sampleB_f=[]
    submit_f=[]
    answer_f=[]
    correct_f=[]
    submit_correct_rate_f=[]
    answer_correct_rate_f=[]
    correct_rate_f=[]
    for i in range(len(table)-12):
        for j in range(1,8):   
            question_f.append(table[i].find_all('tr')[j].find_all('td')[0].find_all('a', href=True)[0].get('href'))
            code_f.append(table[i].find_all('tr')[j].find_all('td')[0].find_all('a', href=True)[2].get('href'))
            sampleA_f.append(table[i].find_all('tr')[j].find_all('td')[1].find_all('a', href=True)[0].get('href'))
            sampleB_f.append(table[i].find_all('tr')[j].find_all('td')[1].find_all('a', href=True)[1].get('href'))
            #for j in range(2,8):
            submit_f.append(table[i].find_all('tr')[j].find_all('td')[2].text)
            answer_f.append(table[i].find_all('tr')[j].find_all('td')[3].text)
            correct_f.append(table[i].find_all('tr')[j].find_all('td')[4].text)
            submit_correct_rate_f.append(table[i].find_all('tr')[j].find_all('td')[5].text)
            answer_correct_rate_f.append(table[i].find_all('tr')[j].find_all('td')[6].text)
            correct_rate_f.append(table[i].find_all('tr')[j].find_all('td')[7].text)
    row_title=[]
    row_title.append('題號/題目')
    row_title.append('程式碼')
    row_title.append('測資a')
    row_title.append('測資b')
    row_title.append('繳交次數')
    row_title.append('作答人數')
    row_title.append('答對人數')
    row_title.append('繳交答對率')
    row_title.append('作答答對率')
    row_title.append('考生答對率')
    activity=pd.DataFrame({row_title[0]:question_f,row_title[1]:code_f,row_title[2]:sampleA_f,row_title[3]:sampleB_f,row_title[4]:submit_f,row_title[5]:answer_f,row_title[6]:correct_f,row_title[7]:submit_correct_rate_f,row_title[8]:answer_correct_rate_f,row_title[9]:correct_rate_f})
    msg = ''
#     for index,row in activity.iterrows(): #根據keyword收尋網頁，列出keyword所需的活動
#         msg += row['日期'] +' ' + row['標題']+'\n'+row['連結']+'\n' 
#     msg.strip()
#     if msg=='':
#         msg = '沒有活動'
    answer=random.randint(0,len(activity))
    
    #msg+=activity[answer:answer+1]
    finalreturn=activity[answer:answer+1].to_json(orient="columns",force_ascii=False).replace('"'+str(answer)+'":','')
    finalreturn=finalreturn.replace('{','')
    finalreturn=finalreturn.replace('}','')
    finalreturn=finalreturn.replace(',','\n')
    finalreturn=finalreturn.replace('"','')
    return finalreturn.replace('\\','')
#    return msg[:500]
```

```python
#沒有時間的
import json
from flask import Flask, request, make_response, jsonify

app = Flask(__name__)


@app.route("/")
def verify():
    return "Hello world", 200


@app.route('/', methods=['POST']) #將其架設到本地端上，5000因我的其他課程使用 改port為3000
def webhook():
    req = request.get_json(silent=True, force=True)  
    print(req)
    keyword=None
    if req['queryResult']['parameters']['any'] != '':
        keyword = req['queryResult']['parameters']['any']
        
    res = {"fulfillmentText": get_activity(keyword=keyword)}
    return make_response(jsonify(res))

if __name__ == '__main__':
    app.run(port=3000)
```
